```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score , confusion_matrix
```

```python
data = pd.read_csv("/content/Iris.csv")
data
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data['Target'] = data['Species'].map({'Iris-setosa':0 , 'Iris-versicolor':1, 'Iris-virginica':2 })
data
```

```python
x = data[['SepalLengthCm', 'SepalWidthCm'  , 'PetalLengthCm' , 'PetalWidthCm']]
y = data['Target']
```

```python
x
```

```python
y
```

```python
xTrain , xTest , yTrain, yTest = train_test_split(x , y , test_size=0.2 , random_state=45)
```

```python
model = RandomForestClassifier(n_estimators=110 , max_depth= None , random_state=45 )

model.fit(xTrain , yTrain)

```

```python
prediction = model.predict(xTest)
```

```python
yTest
```

```python
prediction
```

```python
accuracy = accuracy_score(yTest , prediction)
accuracy
```

```python
confusion = confusion_matrix(yTest , prediction)
confusion
```

```python
sns.heatmap(confusion_matrix(yTest , prediction) ,annot=True, cmap='coolwarm')
plt.title("Confusion Matrix")
plt.show()
```

```python
tuned = RandomForestClassifier(n_estimators=50, max_depth=5, random_state=42)
tuned.fit(xTrain, yTrain)

```

```python
tunedPrediction = tuned.predict(xTest)
tunedPrediction
```

```python
tunedAccuracy = accuracy_score(yTest, tunedPrediction)
tunedAccuracy
```

```python
confusion_matrix(yTest, tunedPrediction)
```

```python
sns.heatmap(confusion_matrix(yTest, tunedPrediction) ,annot=True, cmap='coolwarm')
plt.title("Confusion Matrix After Adjusting Hyperparameters")
plt.show()
```



1.   **Random Forest** is a machine learning algorithm which works multiple decision trees to classify data into different categories and predicts based on maximum votes.
2.   **Advantages:**

*   It prevents from the overfitting as compared to other decisions tree.
*   It works on both Regression and Classifier data
*   It can handle some missing data and work properly even if some data is missing.
*   It is more accurate , especially in large dataset.
